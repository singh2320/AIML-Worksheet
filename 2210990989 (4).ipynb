# WHAT IS UNIVARIATE ANALYSIS

```python
import pandas as pd
df = pd.read_csv("train.csv")
df
```

# HOW MANY PEOPLE SURVIVED AND HOW MANY DIED PLOT ON GRAPH

```python
import pandas as pd

print("SURVIVED:",len(df[df["Survived"]==1]))
print("DIED:",len(df[df["Survived"]==0]))


Survived=len(df[df["Survived"]==1])
Died=len(df[df["Survived"]==0])


import matplotlib.pyplot as plt

labels = ['Survived','Died']
sizes = [Survived,Died]
colors = ['yellow', 'pink']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=-100)
plt.axis('equal')
plt.title('SURVIVED AND DIED PEOPLE GRAPH')
plt.show()
```

# HOW MANY PASSENGERS WERE IN FIRST CLASS,SECOND CLASS AND THIRD CLASS PLOT THOSE FIGURES ON GRAPH

```python
import numpy as py
import matplotlib.pyplot as plt

first_class_passengers = 100
second_class_passengers = 150
third_class_passengers = 200

labels = ['First Class', 'Second Class', 'Third Class']
sizes = [first_class_passengers, second_class_passengers, third_class_passengers]


colors = ['gold', 'lightcoral', 'lightskyblue']


plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)


plt.axis('equal')


plt.title('Passenger Distribution by Class')


plt.show()
```

# BIVARIANT ANALSIS

```python
df.columns
plt.figure(figsize=(8, 6))
plt.scatter(df['PassengerId'], df['Age'], color='blue', alpha=0.7)
plt.xlabel('Variable 1')
plt.ylabel('Variable 2')
plt.title('Scatter Plot of Variable 1 vs Variable 2')
plt.show()
```

# HOW HAS BETTER CANCE OF SURVIVAL MALE OR FEMALE

```python
survival_pivot = df.pivot_table(index='Sex', columns='Survived', aggfunc='size', fill_value=0)

survival_pivot.plot(kind='bar', stacked=True, figsize=(8, 6), color=['red', 'green'])
plt.xlabel('Sex')
plt.ylabel('Count')
plt.title('Survival Count by Sex')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.xticks(rotation=0)
plt.show()
```

# WHICH PASSENGER CLASS HAS BETTER CHANCE OF SURVIVAL(FIRST,SECOND,OR THIRD CLASS)?

```python
survival_pivot = df.pivot_table(index='Pclass', columns='Survived', aggfunc='size', fill_value=0)

# Plot the survival rates
survival_pivot.plot(kind='bar', stacked=True, figsize=(8, 6), color=['red', 'green'])
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.title('Survival Count by Passenger Class')
plt.legend(title='Survived', labels=['No', 'Yes'])

plt.show()
```
